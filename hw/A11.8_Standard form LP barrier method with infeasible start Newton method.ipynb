```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
m = 100
n = 500

np.random.seed(0)

# Generate data matrix
A = np.random.rand(m, n)
b = A @ np.random.rand(n, 1)
c = np.random.rand(n, 1)

x = cp.Variable((n, 1))

c1 = c.T@x
obj = cp.Minimize(c1)
Problem = cp.Problem(obj, [A @x == b, x >= 0])
Problem.solve()

print(f"min cost:{c1.value}")
```
