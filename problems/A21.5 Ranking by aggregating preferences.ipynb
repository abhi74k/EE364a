```python
import cvxpy as cp
import numpy as np
import pandas as pd

preferences, n, m = synt_rank_aggr_data()

def create_ranking(penalty='l2'):
    ranks = cp.Variable(n)

    #objective = cp.sum([cp.square(cp.pos(ranks[rj] + 1 + ranks[ri])) for (ri, rj) in preferences])

    violations = cp.pos(ranks[preferences[:, 1]] - ranks[preferences[:, 0]] + 1)
    
    if penalty == 'l1':
        objective = cp.sum(violations)
    else:
        objective = cp.sum(cp.square(violations))

    problem = cp.Problem(cp.Minimize(objective))
    problem.solve()
    print("Status:", problem.status)

    return ranks.value, violations.value
```
