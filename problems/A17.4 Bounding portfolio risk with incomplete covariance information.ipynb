```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.1, 0.2, -0.05, 0.1])
S = cp.Variable((4, 4), PSD=True)

cost = cp.Maximize(x.T @ S @x)
constraints = [
    S[0][0] == 0.2,
    S[1][1] == 0.1,
    S[2][2] == 0.3,
    S[3][3] == 0.1,
    
    S[0][1] >= 0.0,
    S[0][2] >= 0.0,
    
    S[1][2] <= 0.0,
    S[1][3] <= 0.0,
    
    S[2][3] >= 0.0
]

max_var = cp.Problem(cost, constraints).solve()
print(f'\nMax variance: {np.round(max_var, 4)}')
print(f'\nWorst case co-variance matrix:\n{np.round(S.value, 4)}')
print(f'\nRisk with diagonal elements:{np.round(x.T@np.diag([0.2, 0.1, 0.3, 0.1]) @x, 4)}')
```
