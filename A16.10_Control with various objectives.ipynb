```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 4
m = 2 

A = np.array([
[ 0.95,  0.16,  0.12,  0.01],
[-0.12,  0.98, -0.11, -0.03],
[-0.16,  0.02,  0.98,  0.03],
[-0.  ,  0.02, -0.04,  1.03],
])

B = np.array([
[ 0.8 , 0. ],
[ 0.1 , 0.2],
[ 0.  , 0.8],
[-0.2 , 0.1],
])

x_init = np.ones(n)

T = 100
```

```python
def plot_control(u):
  u1 = u[0, :]
  u2 = u[1, :]

  # Create the figure and subplots
  fig, axs = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 8))
  
  # Plot the first subplot with u1 on the y-axis
  axs[0].plot(u1, label='u1')
  axs[0].plot(u2, label='u2')
  axs[0].set_xlabel('t')

  # Plot the second subplot with u2 on the y-axis
  axs[1].plot(cp.norm2(u, axis=0).value, label='2-norm')
  axs[1].set_xlabel('t')

  axs[0].legend()
  axs[1].legend()
  plt.show()

x_final = np.zeros(n)

X = []
for t in np.arange(T-1, 0, -1):
  X.append(np.linalg.matrix_power(A, t) @ B)
X = np.hstack(X)


u = cp.Variable((2, T-1))

problem = cp.Problem(cp.Minimize(cp.sum_squares(cp.vec(u))), [X @ cp.vec(u) == -np.linalg.matrix_power(A, T) @ x_init])
min_cost1 = problem.solve()
print(f'Sum of squares of 2-norm cost: {min_cost1}')

plot_control(u.value.reshape(2, -1))
```

```python
problem = cp.Problem(cp.Minimize(cp.sum(cp.norm2(u, axis=0))), [X @ cp.vec(u) == -np.linalg.matrix_power(A, T) @ x_init])
min_cost1 = problem.solve()
print(f'Sum of 2-norms cost: {min_cost1}')

plot_control(u.value.reshape(2, -1))
```

```python
problem = cp.Problem(cp.Minimize(cp.max(cp.norm2(u, axis=0))), [X @ cp.vec(u) == -np.linalg.matrix_power(A, T) @ x_init])
min_cost1 = problem.solve()
print(f'Max of 2 norms cost: {min_cost1}')

plot_control(u.value.reshape(2, -1))
```

```python
problem = cp.Problem(cp.Minimize(cp.sum(cp.norm1(u, axis=0))), [X @ cp.vec(u) == (-np.linalg.matrix_power(A, T) @ x_init)])
min_cost1 = problem.solve()
print(f'Sum of 1-norms cost: {min_cost1}')

plot_control(u.value.reshape(2, -1))
```
