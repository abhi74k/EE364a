```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np
from scipy.sparse import random as sprand
m=20
n=10
p=8
np.random.seed(2)
R = np.round(np.random.rand(m, n))
A = sprand(p, m, density=0.2).toarray()
c = 10*np.random.rand(m) + 20
b = 20*np.random.rand(p) + 30
```

```python
def constrained_opt():
  f = cp.Variable(n)

  obj = 0
  for i in np.arange(n):
    obj += cp.sqrt(f[i])
 
  constraints = [
      R @ f <= c,
      A @ R @ f <= b,
      f >= 0
  ]

  problem = cp.Problem(cp.Maximize(obj), constraints)
  cost = problem.solve()
  f_star = f.value

  return cost, f_star

cost, f_star = constrained_opt()
t = R @ f_star


print(f"Optimal cost: {cost}")
print(f"f_star: {f_star}")

eps = 0.001
at_capacity_links = np.isclose(t, c, atol=eps)
print(f"At capacity links : {np.where(at_capacity_links == True)[0] + 1}")
```

```python
def constrained_opt_without_interference():
  f = cp.Variable(n)

  obj = 0
  for i in np.arange(n):
    obj += cp.sqrt(f[i])

  constraints = [
      R @ f <= c, 
      f >= 0
  ]

  problem = cp.Problem(cp.Maximize(obj), constraints)
  cost = problem.solve()
  f_star = f.value

  return cost, f_star

cost, f_star = constrained_opt_without_interference()
t = R @ f_star


print(f"Optimal cost: {cost}")
print(f"f_star: {f_star}")

eps = 0.001
at_capacity_links = np.isclose(t, c, atol=eps)
print(f"At capacity links : {np.where(at_capacity_links == True)[0] + 1}")
```
