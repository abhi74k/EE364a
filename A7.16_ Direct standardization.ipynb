```python
p  = cp.Variable([N, 1])

obj = cp.Maximize(cp.sum(cp.entr(p)))
constraints = [
    p >= 0, 
    cp.sum(p) == 1,
    x.T @ p == Ex
]
prob = cp.Problem(obj, constraints)
prob.solve()

p = p.value

p_est = []
for i in np.arange(1, 21):
  p_est.append(np.sum(p[y == i]))

print(f'Weighted empirical distribution:\n')
p_est
```
